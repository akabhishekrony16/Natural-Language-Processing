```python
from nltk.corpus import gutenberg
gutenberg.fileids()
```

```python
emma = gutenberg.words('austen-emma.txt')
len(emma)
```

```python
#computing statistics for the gutenberg field
for fileid in gutenberg.fileids():
    num_chars = len(gutenberg.raw(fileid))#getting number of letters in text including spaces
    num_words = len(gutenberg.words(fileid))#number of words
    num_sents = len(gutenberg.sents(fileid))#number of sentences
    num_vocab = len(set([w.lower() for w in gutenberg.words(fileid)]))
    print (int(num_chars/num_words), int(num_words/num_sents), int(num_words/num_vocab),fileid)
```

### Brown Corpus

```python
from nltk.corpus import brown
brown.categories()
```

```python
brown.words(categories='news')
```

```python
brown.words(fileids=['cg22'])
```

```python
news_text = brown.words(categories='news')
```

```python
fdist = nltk.FreqDist([w.lower() for w in news_text])
```

```python
modals = ['can', 'could', 'may', 'might', 'must', 'will']
for m in modals:
    print (m + ':', fdist[m])
```

```python
adventure_text = brown.words(categories='adventure')
modals = ['what','when', 'where', 'who','why']
fdist = nltk.FreqDist([w.lower() for w in adventure_text])
for m in modals:
    print (m + ':', fdist[m])
```

```python
cfd = nltk.ConditionalFreqDist((genre, word)for genre in brown.categories()for word in brown.words(categories=genre))
genres = ['news', 'religion', 'hobbies', 'science_fiction', 'romance', 'humor']
modals = ['can', 'could', 'may', 'might', 'must', 'will']
cfd.tabulate(conditions=genres, samples=modals)
```
