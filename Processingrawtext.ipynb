```python
import nltk, re, pprint
```

### Regular Expressions for Detecting Word Patterns

```python
wordlist = [w for w in nltk.corpus.words.words('en') if w.islower()]
```

#### Using Basic Metacharacters

Let’s find words ending with ed using the regular expression «ed$». We will use the re.search(p, s) function to check whether the pattern p can be found somewhere inside the string s. We need to specify the characters of interest, and use the dollar sign,which has a special behavior in the context of regular expressions in that it matches theend of the word:

```python
[w for w in wordlist if re.search('ed$', w)]
```

```python
[w for w in wordlist if re.search('^..j..t..$', w)]
```

##### The caret symbol ^ matches the start of a string, just like the $ matches the end.

```python
[w for w in wordlist if re.search('..j..t..', w)]
```

#### Finally, the ? symbol specifies that the previous character is optional. Thus «^e-?mail $» will match both email and e-mail.

```python
sum(1 for w in wordlist if re.search('^e-?mail$', w))
```

### Ranges and Closures

```python
[w for w in wordlist if re.search('^[ghi][mno][jlk][def]$', w)]#textonyms
```

```python
[w for w in wordlist if re.search('^[ghijklmno]+$', w)]#textonyms
```

```python
[w for w in wordlist if re.search('^[g-o]+$', w)]#textonyms
```

```python
[w for w in wordlist if re.search('^[a-fj-o]+$', w)]#textonyms
```

### Basic regular expression metacharacters, including wildcards, ranges, and closures

```python
chat_words = sorted(set(w for w in nltk.corpus.nps_chat.words()))
[w for w in chat_words if re.search('^m+i+n+e+$', w)]
```

```python
[w for w in chat_words if re.search('^[ha]+$', w)]
```

```python
[w for w in chat_words if re.search('^m*i*n*e*$', w)]
```

```python
[w for w in chat_words if re.search('^[^aeiouAEIOU]+$', w)]#matches any character other than a vowel
```

```python
wsj = sorted(set(nltk.corpus.treebank.words()))
```

```python
[w for w in wsj if re.search('^[0-9]+\.[0-9]+$', w)]
```

```python
[w for w in wsj if re.search('^[A-Z]+\$$', w)]
```

```python
[w for w in wsj if re.search('^[0-9]{4}$', w)]
```

```python
[w for w in wsj if re.search('^[0-9]+-[a-z]{3,5}$', w)]
```

```python
[w for w in wsj if re.search('^[a-z]{5,}-[a-z]{2,3}-[a-z]{,6}$', w)]
```

```python
[w for w in wsj if re.search('(ed|ing)$', w)]
```
