```python
def unknown(url):
    html = urlopen(url).read()
    soup = BeautifulSoup(html, 'html.parser')
    tokens = nltk.word_tokenize(soup.get_text())
    A = set([w for w in tokens if re.findall(r'^[a-z]+$', w)])
    B = set([w for w in nltk.corpus.words.words('en') if w.islower()])
    return list(A-B)
```

```python
unknown("http://news.bbc.co.uk/2/hi/health/2284783.stm")
```

```python
sum([len(i) for i in words])/len([len(i) for i in words])
```

```python
def ARI(category):
    words = [w for w in brown.words(categories=category) if re.search('\w',w)]
    mu_w = sum([len(i) for i in words])/len([len(i) for i in words])
    sen = [[w for w in sentences if re.search('\w',w)]for sentences in brown.sents(categories=category)]
    mu_s = sum([len(i) for i in sen])/len([len(i) for i in sen])
    return 4.71*mu_w + 0.5*mu_s -21.43
```

```python
ARI('adventure')
```

```python
for i in brown.categories():
    print(ARI(i))
```
